```python
import pandas as pd
from pandas_profiling import ProfileReport
#import xlrd
```

## Dataset Exploration

### UCI Bank Marketing Dataset (https://archive.ics.uci.edu/dataset/222/bank+marketing)

There are two versions: one with ~17 features and this one with 21.

```python
df = pd.read_csv('data/bank-additional/bank-additional-full.csv', sep=';', header=0)
df.head()
```

```python
df.info()
```

```python
print(df.describe())
print(df.describe(include='object'))
```

```python
profile = ProfileReport(df, title="Pandas Profiling Report")
#profile.to_notebook_iframe()
profile.to_file("bank-additional-full.html")
```

### CDC Diabetes Health Indicators (https://archive.ics.uci.edu/dataset/891/cdc+diabetes+health+indicators)

There are three versions of the dataset. Two that have binary responses (either balanced or unbalanced) and this one, unbalanced with three response categories.

```python
df = pd.read_csv('data/diabetes/diabetes_012_health_indicators_BRFSS2015.csv', header=0)
df.head()
```

```python
df.info()
```

```python
print(df.describe())
```

```python
profile = ProfileReport(df, title="Pandas Profiling Report")
#profile.to_notebook_iframe()
profile.to_file("diabetes.html")
```
